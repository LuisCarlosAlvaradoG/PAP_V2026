## Funciones de los modelos

```python
def obtener_proyeccion(df, horizonte):
    """
    Proyección adaptativa de ingresos y egresos semanales.

    Estrategia automática según la cantidad de semanas históricas:
        < 8 semanas       : promedio simple de todo el historial.
        8 a 51 semanas    : media móvil ponderada de las últimas 8 semanas.
        >= 52 semanas     : media móvil ponderada + factor estacional por
                            semana del año (ISO week).

    Siempre devuelve dos listas de longitud `horizonte`:
        I : ingresos proyectados por semana
        E : egresos proyectados por semana
    """
    # --------------------------------------------------------------
    # 1. Agregar histórico a nivel semanal (mismo método que el original)
    # --------------------------------------------------------------
    resumen = df.pivot_table(
        index="inicio_semana", columns="tipo", values="importe",
        aggfunc="sum", fill_value=0,
    ).sort_index()

    for col in ["Ingreso", "Egreso"]:
        if col not in resumen.columns:
            resumen[col] = 0.0

    ingreso_hist = resumen["Ingreso"].values.astype(float)
    egreso_hist  = resumen["Egreso"].values.astype(float)
    n_semanas = len(ingreso_hist)          # semanas con al menos un movimiento

    # --------------------------------------------------------------
    # 2. Pesos para media móvil ponderada (últimas 8 semanas)
    #    Pesos lineales crecientes: la más reciente pesa más.
    # --------------------------------------------------------------
    pesos = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=float)
    pesos = pesos / pesos.sum()            # normalizados

    def wma(serie):
        """Media móvil ponderada de las últimas 8 semanas (o menos)."""
        ultimas = serie[-8:]               # toma hasta 8 elementos
        n = len(ultimas)
        # Ajustar pesos a la cantidad real de semanas disponibles
        pesos_ajustados = pesos[-n:]       # últimos n pesos
        pesos_ajustados = pesos_ajustados / pesos_ajustados.sum()
        return np.dot(ultimas, pesos_ajustados)

    # --------------------------------------------------------------
    # 3. Factores estacionales (solo si hay >= 52 semanas)
    # --------------------------------------------------------------
    factores_ingreso = None
    factores_egreso = None
    if n_semanas >= 52:
        # Obtener número de semana ISO para cada fecha del histórico
        semanas_iso = resumen.index.isocalendar().week.values

        def calcular_factores(serie):
            df_temp = pd.DataFrame({"semana": semanas_iso, "valor": serie})
            promedio_semana = df_temp.groupby("semana")["valor"].mean()
            global_avg = serie.mean()
            if global_avg == 0:
                # Si no hay ingresos/egresos, factor 1 para todo
                return {s: 1.0 for s in range(1, 54)}
            return (promedio_semana / global_avg).to_dict()

        factores_ingreso = calcular_factores(ingreso_hist)
        factores_egreso  = calcular_factores(egreso_hist)

    # --------------------------------------------------------------
    # 4. Proyección según el caso
    # --------------------------------------------------------------
    if n_semanas == 0:
        # Sin historial → devolver ceros (robustez)
        return [0.0] * horizonte, [0.0] * horizonte

    # Fechas futuras: a partir del último inicio_semana registrado
    ultima_fecha = resumen.index[-1]
    fechas_futuras = pd.date_range(
        start=ultima_fecha + pd.Timedelta(weeks=1),
        periods=horizonte,
        freq="W"
    )
    semanas_futuras = fechas_futuras.isocalendar().week

    def proyectar(serie_hist, factores):
        if n_semanas < 8:
            # Caso 1: promedio simple de todo el historial
            base = serie_hist.mean()
            return np.full(horizonte, base)

        # Caso 2 y 3: media móvil ponderada de las últimas 8
        base = wma(serie_hist)

        if n_semanas < 52 or factores is None:
            # Caso 2: sin estacionalidad
            return np.full(horizonte, base)

        # Caso 3: con estacionalidad
        proy = np.array([base * factores.get(s, 1.0) for s in semanas_futuras])
        return proy

    I = proyectar(ingreso_hist, factores_ingreso).tolist()
    E = proyectar(egreso_hist, factores_egreso).tolist()

    return I, E

def modelo3_resiliencia(
    I, E, C0, T,
    R, F, s,
    E_max=None,
    L_min=None,
    D0=None   # ignorado, solo por uniformidad
):
    """
    Modelo 3 – Resiliencia pura del flujo de caja.
    Maximiza el piso de caja reprogramando los egresos flexibles
    dentro de una ventana temporal ±s.
    """
    I = np.asarray(I, dtype=float)
    E = np.asarray(E, dtype=float)
    R = np.asarray(R, dtype=float)
    F = np.asarray(F, dtype=float)
    if not (len(I) == len(E) == len(R) == len(F) == T):
        raise ValueError("I, E, R, F deben tener longitud T")
    if np.any(F < 0) or np.any(R < 0):
        raise ValueError("R y F deben ser no negativos")

    # Si no hay flexibilidad, simulación directa
    if np.all(F == 0):
        C_sim = np.zeros(T+1)
        C_sim[0] = C0
        for t in range(1, T+1):
            C_sim[t] = C_sim[t-1] + I[t-1] - E[t-1]
        m_opt = np.min(C_sim[1:])
        return dict(
            m_opt=m_opt,
            C_opt=C_sim.tolist(),
            E_opt=E.tolist(),
            x_opt={},
            exito=True,
            t_crit=int(np.argmin(C_sim[1:])) + 1,
            K_star=(np.max(E) / np.sum(E)) if np.sum(E) > 0 else np.nan
        )

    # Construir variables: solo pares (τ,t) con |t-τ| ≤ s.
    pairs = []          # lista de (tau, t) en orden
    idx_map = {}        # (tau, t) -> índice de variable
    n_x = 0
    for tau in range(T):
        t_start = max(0, tau - s)
        t_end = min(T, tau + s + 1)
        for t in range(t_start, t_end):
            pairs.append((tau, t))
            idx_map[(tau, t)] = n_x
            n_x += 1

    n_vars = n_x + 1
    idx_m = n_x  # índice de m

    # Función objetivo: maximizar m  =>  minimizar -m
    c = np.zeros(n_vars)
    c[idx_m] = -1.0

    # Restricciones de igualdad: ∑_t x_{τ,t} = F_τ
    A_eq = []
    b_eq = []
    for tau in range(T):
        row = np.zeros(n_vars)
        t_start = max(0, tau - s)
        t_end = min(T, tau + s + 1)
        for t in range(t_start, t_end):
            row[idx_map[(tau, t)]] = 1.0
        A_eq.append(row)
        b_eq.append(F[tau])
    A_eq = np.array(A_eq)
    b_eq = np.array(b_eq)

    # Restricciones de desigualdad
    A_ub = []
    b_ub = []

    # Dinámica + Piso: C_t >= m para t=1..T
    cum_I = np.cumsum(I)
    cum_R = np.cumsum(R)
    for t_idx in range(1, T+1):
        row = np.zeros(n_vars)
        for k in range(t_idx):
            tau_start = max(0, k - s)
            tau_end = min(T, k + s + 1)
            for tau in range(tau_start, tau_end):
                if (tau, k) in idx_map:
                    row[idx_map[(tau, k)]] -= 1.0
        row[idx_m] = -1.0
        rhs = -(C0 + cum_I[t_idx-1] - cum_R[t_idx-1])
        A_ub.append(row)
        b_ub.append(rhs)

    # Capacidad semanal opcional
    if E_max is not None:
        E_max = np.asarray(E_max, dtype=float)
        if len(E_max) != T:
            raise ValueError("E_max debe tener longitud T")
        for t in range(T):
            if np.isfinite(E_max[t]):
                row = np.zeros(n_vars)
                tau_start = max(0, t - s)
                tau_end = min(T, t + s + 1)
                for tau in range(tau_start, tau_end):
                    row[idx_map[(tau, t)]] = 1.0
                rhs = E_max[t] - R[t]
                A_ub.append(row)
                b_ub.append(rhs)

    # Colchón de liquidez: m ≥ L_min
    if L_min is not None:
        row = np.zeros(n_vars)
        row[idx_m] = -1.0   # -m <= -L_min
        A_ub.append(row)
        b_ub.append(-L_min)

    A_ub = np.array(A_ub) if A_ub else None
    b_ub = np.array(b_ub) if b_ub else None

    bounds = [(0, None)] * n_x + [(None, None)]

    res = linprog(c, A_ub=A_ub, b_ub=b_ub, A_eq=A_eq, b_eq=b_eq,
                  bounds=bounds, method='highs')

    if not res.success:
        return dict(m_opt=np.nan, C_opt=[], E_opt=[],
                    x_opt={}, exito=False, t_crit=None, K_star=np.nan)

    x_sol = res.x[:n_x]
    m_opt = res.x[idx_m]

    x_opt = {pair: x_sol[idx_map[pair]] for pair in pairs}

    E_opt = np.copy(R)
    for (tau, t), val in x_opt.items():
        E_opt[t] += val

    C_opt = np.zeros(T+1)
    C_opt[0] = C0
    for t in range(1, T+1):
        C_opt[t] = C_opt[t-1] + I[t-1] - E_opt[t-1]

    t_crit = int(np.argmin(C_opt[1:])) + 1
    sum_E_opt = np.sum(E_opt)
    K_star = np.max(E_opt) / sum_E_opt if sum_E_opt > 0 else np.nan

    return dict(m_opt=m_opt, C_opt=C_opt.tolist(),
                E_opt=E_opt.tolist(), x_opt=x_opt,
                exito=True, t_crit=t_crit, K_star=K_star)
def construir_tabla_modelo3(res):
    if not res['exito']:
        return pd.DataFrame({'Error': ['Optimización fallida']})
    df = pd.DataFrame({
        'Semana': list(range(len(res['C_opt']))),
        'Caja': res['C_opt'],
        'Egreso_reprogramado': [res['E_opt'][t-1] if t>0 else np.nan for t in range(len(res['C_opt']))]
    })
    metricas = pd.DataFrame({
        'Semana': ['---', 'Piso máximo (m*)', 'Semana crítica', 'Índice K*'],
        'Caja': ['', res['m_opt'], res['t_crit'], res['K_star']],
        'Egreso_reprogramado': ['', '', '', '']
    })
    return pd.concat([df, metricas], ignore_index=True)

def modelo2_optimizacion_global(D0, I, E, C0, L_min, T, sigma=None, I_medio=None, k=1.0, eps=1e-6):
    """
    Modelo 2 - Optimización convexa global del horizonte completo.
    
    Objetivo:
        min  α·Σ_t max(0, L_min - C_t)²  +  Σ_i w_i · D_{i,T}
    
    Donde:
        - w_i ∝ 1/D0_i (peso mayor a deuda más pequeña → prioridad snowball suave)
        - α = (1 / L_min²) · (1 + k·(σ / I_medio))  si se proporciona volatilidad.
    
    Parámetros:
        D0      : lista/array de deudas iniciales (longitud n)
        I, E    : listas de ingresos y egresos semanales (longitud T)
        C0      : caja inicial
        L_min   : colchón de liquidez deseado
        T       : horizonte en semanas
        sigma   : volatilidad histórica de los flujos netos (opcional)
        I_medio : ingreso semanal medio para normalizar volatilidad (opcional)
        k       : sensibilidad a la volatilidad (default 1)
        eps     : constante para estabilidad numérica
    
    Retorna:
        P      : matriz numpy (n x T) con pagos planeados
        ratios : lista de longitud T con el indicador deuda/(caja+L_min)
        exito  : bool, True si se encontró solución óptima
    """
    n = len(D0)
    
    # ---- Calcular coeficiente α ----
    if L_min > 0:
        alpha_base = 1.0 / (L_min ** 2)
    else:
        escala = max(C0, sum(D0), 1.0)
        alpha_base = 1.0 / (escala ** 2)
    if sigma is not None and I_medio is not None and I_medio > 0:
        alpha = alpha_base * (1.0 + k * (sigma / I_medio))
    else:
        alpha = alpha_base

    # ---- Pesos snowball: mayor peso a deuda pequeña ----
    max_D0 = max(D0)
    w = np.array([max_D0 / (d + eps) for d in D0])

    # ---- Variables de decisión ----
    P = cp.Variable((n, T), nonneg=True)   # pagos
    C = cp.Variable(T+1)                   # caja al final de cada semana
    D = cp.Variable((n, T+1))              # deuda pendiente

    # ---- Restricciones ----
    constraints = [C[0] == C0, D[:, 0] == D0]
    for t in range(1, T+1):
        idx = t-1
        # Caja
        constraints.append(C[t] == C[t-1] + I[idx] - E[idx] - cp.sum(P[:, t-1]))
        # Deuda
        constraints.append(D[:, t] == D[:, t-1] - P[:, t-1])
        constraints.append(D[:, t] >= 0)
        # No quiebra
        constraints.append(C[t] >= 0)

    # ---- Función objetivo ----
    deficit = cp.pos(L_min - C[1:])            # máx(0, L_min - C_t)
    coste_liquidez = alpha * cp.sum_squares(deficit)
    coste_deuda = w @ D[:, T]
    objetivo = cp.Minimize(coste_liquidez + coste_deuda)

    # ---- Resolver (QP) ----
    prob = cp.Problem(objetivo, constraints)
    prob.solve(solver=cp.OSQP)                 # o cp.ECOS

    if prob.status not in ["optimal", "optimal_inaccurate"]:
        return np.zeros((n, T)), [0.0]*T, False

    P_opt = P.value
    C_opt = C.value
    D_opt = D.value

    # ---- Ratios D/(C+L_min) para diagnóstico ----
    ratios = []
    for t in range(1, T+1):
        deuda_t = np.sum(D_opt[:, t])
        caja_t = C_opt[t]
        denom = caja_t + L_min
        if denom <= eps:
            ratio = deuda_t / eps if deuda_t > 0 else 0.0
        else:
            ratio = deuda_t / denom
        ratios.append(float(ratio))

    return P_opt, ratios, True

# ============================================================
#  Carga de datos (igual que en el dashboard original)
# ============================================================
ARCHIVO_DATOS = "../Datos flujo de efectivo.xlsx"
ARCHIVO_PAGOS = "../Pagos pendientes.xlsx"

COLUMNAS_PAGOS = [
    "concepto", "categoría", "tipo", "importe", "semana",
    "inicio_semana", "fecha_registro", "estatus", "fecha_pago",
]

def cargar_datos_flujo(ruta=ARCHIVO_DATOS):
    df = pd.read_excel(ruta)
    df["inicio_semana"] = pd.to_datetime(df["inicio_semana"], dayfirst=True)
    df["tipo"] = df["tipo"].astype(str).str.strip()
    df["categoría"] = df["categoría"].astype(str).str.strip()
    df["descripción"] = df["descripción"].astype(str).str.strip()
    df["importe"] = pd.to_numeric(df["importe"], errors="coerce").fillna(0)
    return df

def cargar_pagos_pendientes(ruta=ARCHIVO_PAGOS):
    try:
        dfp = pd.read_excel(ruta)
        for col in COLUMNAS_PAGOS:
            if col not in dfp.columns:
                dfp[col] = ""
        dfp = dfp[COLUMNAS_PAGOS]
    except FileNotFoundError:
        dfp = pd.DataFrame(columns=COLUMNAS_PAGOS)
    return dfp

def obtener_deudas_pendientes(dfp):
    pend = dfp[dfp["estatus"] == "Pendiente"].copy()
    pend["importe"] = pd.to_numeric(pend["importe"], errors="coerce").fillna(0)
    resumen = pend.groupby("concepto")["importe"].sum()
    return resumen[resumen > 0].sort_values(ascending=False)

def calcular_saldo_actual(df):
    ingresos = df.loc[df["tipo"] == "Ingreso", "importe"].sum()
    egresos = df.loc[df["tipo"] == "Egreso", "importe"].sum()
    return float(ingresos - egresos)

# ============================================================
#  Proyección de flujo de efectivo (versión mejorada)
# ============================================================
def obtener_proyeccion(df, horizonte):
    """
    Proyección adaptativa de ingresos y egresos semanales.
    Estrategia automática según la cantidad de semanas históricas:
        < 8 semanas       : promedio simple de todo el historial.
        8 a 51 semanas    : media móvil ponderada de las últimas 8 semanas.
        >= 52 semanas     : media móvil ponderada + factor estacional por
                            semana del año (ISO week).
    Siempre devuelve dos listas de longitud `horizonte`:
        I : ingresos proyectados por semana
        E : egresos proyectados por semana
    """
    resumen = df.pivot_table(
        index="inicio_semana", columns="tipo", values="importe",
        aggfunc="sum", fill_value=0,
    ).sort_index()

    for col in ["Ingreso", "Egreso"]:
        if col not in resumen.columns:
            resumen[col] = 0.0

    ingreso_hist = resumen["Ingreso"].values.astype(float)
    egreso_hist  = resumen["Egreso"].values.astype(float)
    n_semanas = len(ingreso_hist)

    # Pesos lineales crecientes para la media móvil ponderada (últimas 8 semanas)
    pesos = np.array([1, 2, 3, 4, 5, 6, 7, 8], dtype=float)
    pesos = pesos / pesos.sum()

    def wma(serie):
        ultimas = serie[-8:]
        n = len(ultimas)
        pesos_ajustados = pesos[-n:]
        pesos_ajustados = pesos_ajustados / pesos_ajustados.sum()
        return np.dot(ultimas, pesos_ajustados)

    factores_ingreso = None
    factores_egreso = None
    if n_semanas >= 52:
        semanas_iso = resumen.index.isocalendar().week.values
        def calcular_factores(serie):
            df_temp = pd.DataFrame({"semana": semanas_iso, "valor": serie})
            promedio_semana = df_temp.groupby("semana")["valor"].mean()
            global_avg = serie.mean()
            if global_avg == 0:
                return {s: 1.0 for s in range(1, 54)}
            return (promedio_semana / global_avg).to_dict()
        factores_ingreso = calcular_factores(ingreso_hist)
        factores_egreso  = calcular_factores(egreso_hist)

    if n_semanas == 0:
        return [0.0] * horizonte, [0.0] * horizonte

    ultima_fecha = resumen.index[-1]
    fechas_futuras = pd.date_range(
        start=ultima_fecha + pd.Timedelta(weeks=1),
        periods=horizonte, freq="W"
    )
    semanas_futuras = fechas_futuras.isocalendar().week

    def proyectar(serie_hist, factores):
        if n_semanas < 8:
            base = serie_hist.mean()
            return np.full(horizonte, base)
        base = wma(serie_hist)
        if n_semanas < 52 or factores is None:
            return np.full(horizonte, base)
        proy = np.array([base * factores.get(s, 1.0) for s in semanas_futuras])
        return proy

    I = proyectar(ingreso_hist, factores_ingreso).tolist()
    E = proyectar(egreso_hist, factores_egreso).tolist()
    return I, E

# ============================================================
#  MODELO 5 — OPCIÓN A (Heurística mejorada)
# ============================================================
def modelo5_opcion_a(D0, I, E, C0, L_min, T, agresividad=0.5, epsilon=1e-6):
    n = len(D0)
    IE = np.array(I, dtype=float) - np.array(E, dtype=float)
    D_restante = np.array(D0, dtype=float)
    orden_prioridad = np.argsort(D_restante)  # deudas más pequeñas primero
    C = C0
    P = np.zeros((n, T))
    ratios = []
    for t in range(T):
        C_disponible = C + IE[t]
        excedente = max(0.0, C_disponible - L_min)
        pago_total = min(excedente * agresividad, D_restante.sum())
        if pago_total > 0 and D_restante.sum() > 0:
            restante_a_pagar = pago_total
            for i in orden_prioridad:
                if restante_a_pagar <= 0:
                    break
                abono = min(D_restante[i], restante_a_pagar)
                P[i, t] = abono
                D_restante[i] -= abono
                restante_a_pagar -= abono
        C = C_disponible - P[:, t].sum()
        ratio_t = D_restante.sum() / (C + L_min + epsilon)
        ratios.append(ratio_t)
    return P, ratios, True

# ============================================================
#  MODELO 5 — CFO PRUDENTE (heurística adaptativa)
# ============================================================
def modelo5_cfo_prudente(D0, I, E, C0, L_min, T,
                         L_survival=0.0, D_crit=5.0, piso_agres=0.05, epsilon=1e-6):
    n = len(D0)
    IE = np.array(I, dtype=float) - np.array(E, dtype=float)
    D_restante = np.array(D0, dtype=float)
    C = C0
    P = np.zeros((n, T))
    ratios = []
    for t in range(T):
        C_disp = C + IE[t]
        # indicador de liquidez
        liq_ratio = (C_disp + L_min) / (L_min + epsilon)
        liq_ratio = np.clip(liq_ratio, 0.0, 2.0)
        stress_liq = max(0.0, 1.0 - liq_ratio)
        # indicador de carga de deuda
        denom = max(C_disp + L_min, epsilon)
        if C_disp + L_min <= 0:
            stress_debt = 1.0
        else:
            debt_ratio = D_restante.sum() / denom
            stress_debt = min(debt_ratio / D_crit, 1.0)
        s = max(stress_liq, stress_debt)
        agresividad = max(1.0 - s, piso_agres)
        L_dyn = L_survival + (L_min - L_survival) * (1.0 - s)
        excedente = max(0.0, C_disp - L_dyn)
        pago_total = min(excedente * agresividad, D_restante.sum())
        if pago_total > 0:
            orden = np.argsort(D_restante)
            restante = pago_total
            for i in orden:
                if restante <= 0:
                    break
                abono = min(D_restante[i], restante)
                P[i, t] = abono
                D_restante[i] -= abono
                restante -= abono
        C = C_disp - P[:, t].sum()
        ratio_t = D_restante.sum() / (C + L_min + epsilon)
        ratios.append(ratio_t)
    return P, ratios, True

# ============================================================
#  Funciones de presentación (tablas)
# ============================================================
def construir_tabla_modelo5(P, ratios, conceptos, I, E, C0, L_min):
    n, T = P.shape
    IE = np.array(I, dtype=float) - np.array(E, dtype=float)
    pagos_sem = P.sum(axis=0)
    C = C0 + np.cumsum(IE) - np.cumsum(pagos_sem)
    U = IE - pagos_sem
    filas = []
    for t in range(T):
        fila = {"Semana": t + 1}
        for i, concepto in enumerate(conceptos):
            fila[f"Pagar {concepto} ($)"] = round(float(P[i, t]), 2)
        fila["Utilidad/Retenida ($)"] = round(float(U[t]), 2)
        fila["Caja Proyectada ($)"] = round(float(C[t]), 2)
        fila["Ratio D/C"] = round(float(ratios[t]), 4)
        if C[t] <= L_min:
            estado = "Sin margen"
        elif C[t] <= L_min * 1.5:
            estado = "Ajustado"
        else:
            estado = "Operable"
        fila["Estado"] = estado
        filas.append(fila)
    return pd.DataFrame(filas)
```

## Carga de datos


```python
def cargar_datos_flujo(ruta=ARCHIVO_DATOS):
    """Carga el histórico de movimientos (mismo formato que usa generar_flujo en el dashboard)."""
    df = pd.read_excel(ruta)
    df["inicio_semana"] = pd.to_datetime(df["inicio_semana"], dayfirst=True)
    df["tipo"] = df["tipo"].astype(str).str.strip()
    df["categoría"] = df["categoría"].astype(str).str.strip()
    df["descripción"] = df["descripción"].astype(str).str.strip()
    df["importe"] = pd.to_numeric(df["importe"], errors="coerce").fillna(0)
    return df


def cargar_pagos_pendientes(ruta=ARCHIVO_PAGOS):
    """Carga pagos pendientes. Si el archivo no existe, retorna un DataFrame vacío
    con las columnas esperadas (igual que cargar_pagos() en el dashboard)."""
    try:
        dfp = pd.read_excel(ruta)
        for col in COLUMNAS_PAGOS:
            if col not in dfp.columns:
                dfp[col] = ""
        dfp = dfp[COLUMNAS_PAGOS]
    except FileNotFoundError:
        dfp = pd.DataFrame(columns=COLUMNAS_PAGOS)
    return dfp


def obtener_deudas_pendientes(dfp):
    """Agrupa la deuda pendiente por concepto/proveedor (D_i). Igual que en el dashboard."""
    pend = dfp[dfp["estatus"] == "Pendiente"].copy()
    pend["importe"] = pd.to_numeric(pend["importe"], errors="coerce").fillna(0)
    resumen = pend.groupby("concepto")["importe"].sum()
    return resumen[resumen > 0].sort_values(ascending=False)


def calcular_saldo_actual(df):
    ingresos = df.loc[df["tipo"] == "Ingreso", "importe"].sum()
    egresos = df.loc[df["tipo"] == "Egreso", "importe"].sum()
    return float(ingresos - egresos)


def obtener_proyeccion_base(df, horizonte):
    """Proyecta ingresos/egresos futuros usando el promedio de las últimas 6 semanas
    registradas. Reescrito con pivot_table para evitar el FutureWarning de
    groupby().apply() que tiene la versión del dashboard."""
    resumen = df.pivot_table(
        index="inicio_semana", columns="tipo", values="importe",
        aggfunc="sum", fill_value=0,
    ).sort_index()

    for col in ["Ingreso", "Egreso"]:
        if col not in resumen.columns:
            resumen[col] = 0.0

    ultimas = resumen.tail(6)
    ingreso_prom = float(ultimas["Ingreso"].mean()) if not ultimas.empty else 0.0
    egreso_prom = float(ultimas["Egreso"].mean()) if not ultimas.empty else 0.0
    return [ingreso_prom] * horizonte, [egreso_prom] * horizonte
```

```python
# ============================================================
#  CARGA REAL DE DATOS
# ============================================================
df = cargar_datos_flujo()
dfp = cargar_pagos_pendientes()

deudas = obtener_deudas_pendientes(dfp)
saldo_actual = calcular_saldo_actual(df)

print(f"Movimientos históricos cargados: {len(df)}")
print(f"Pagos pendientes registrados: {len(dfp)}")
print(f"Deudas pendientes por concepto (D_i):")
display(deudas)
print(f"\nSaldo actual (C0): {saldo_actual:,.2f}")
```

```python
# ============================================================
#  PARÁMETROS DE PRUEBA (ajustar libremente en cada corrida)
# ============================================================
HORIZONTE = 8          # T, semanas de planeación
L_MIN = round(max(saldo_actual * 0.1, 0.0), 2)   # caja mínima deseada

I, E = obtener_proyeccion_base(df, HORIZONTE)
C0 = saldo_actual
conceptos = deudas.index.tolist()
D0 = deudas.values.tolist()

print(f"Horizonte: {HORIZONTE} semanas")
print(f"L_min: {L_MIN:,.2f}")
print(f"Ingreso proyectado/semana: {I[0]:,.2f}")
print(f"Egreso proyectado/semana: {E[0]:,.2f}")
print(f"C0: {C0:,.2f}")
print(f"Conceptos con deuda: {conceptos}")
print(f"D0: {[round(d,2) for d in D0]}")
```


## Modelo 5 Mínimo ratio deuda/caja


```python
def modelo5_opcion_a(D0, I, E, C0, L_min, T, agresividad=0.5, epsilon=1e-6):
    n = len(D0)
    IE = np.array(I, dtype=float) - np.array(E, dtype=float)
    D_restante = np.array(D0, dtype=float)
    orden_prioridad = np.argsort(D_restante)  # deudas más pequeñas primero (estático)
    C = C0
    P = np.zeros((n, T))
    ratios = []
    for t in range(T):
        C_disponible = C + IE[t]
        excedente = max(0.0, C_disponible - L_min)
        pago_total = min(excedente * agresividad, D_restante.sum())
        if pago_total > 0 and D_restante.sum() > 0:
            restante_a_pagar = pago_total
            for i in orden_prioridad:
                if restante_a_pagar <= 0:
                    break
                abono = min(D_restante[i], restante_a_pagar)
                P[i, t] = abono
                D_restante[i] -= abono
                restante_a_pagar -= abono
        C = C_disponible - P[:, t].sum()
        ratio_t = D_restante.sum() / (C + L_min + epsilon)
        ratios.append(ratio_t)
    return P, ratios, True

def modelo5_cfo_prudente(D0, I, E, C0, L_min, T,
                         L_survival=0.0, D_crit=5.0, piso_agres=0.05, epsilon=1e-6):
    n = len(D0)
    IE = np.array(I, dtype=float) - np.array(E, dtype=float)
    D_restante = np.array(D0, dtype=float)
    C = C0
    P = np.zeros((n, T))
    ratios = []
    for t in range(T):
        C_disp = C + IE[t]
        # Indicador de liquidez (0 = holgado, 1 = crítico)
        liq_ratio = (C_disp + L_min) / (L_min + epsilon)
        liq_ratio = np.clip(liq_ratio, 0.0, 2.0)
        stress_liq = max(0.0, 1.0 - liq_ratio)
        # Indicador de carga de deuda
        denom = max(C_disp + L_min, epsilon)
        if C_disp + L_min <= 0:
            stress_debt = 1.0
        else:
            debt_ratio = D_restante.sum() / denom
            stress_debt = min(debt_ratio / D_crit, 1.0)
        # Estrés global (el peor de los dos)
        s = max(stress_liq, stress_debt)
        # Agresividad adaptativa (con piso mínimo)
        agresividad = max(1.0 - s, piso_agres)
        # Colchón dinámico: en estrés se reduce hasta L_survival
        L_dyn = L_survival + (L_min - L_survival) * (1.0 - s)
        excedente = max(0.0, C_disp - L_dyn)
        pago_total = min(excedente * agresividad, D_restante.sum())
        if pago_total > 0:
            orden = np.argsort(D_restante)      # Snowball dinámico
            restante = pago_total
            for i in orden:
                if restante <= 0:
                    break
                abono = min(D_restante[i], restante)
                P[i, t] = abono
                D_restante[i] -= abono
                restante -= abono
        C = C_disp - P[:, t].sum()
        ratio_t = D_restante.sum() / (C + L_min + epsilon)
        ratios.append(ratio_t)
    return P, ratios, True
```

```python
def construir_tabla_modelo5(P, ratios, conceptos, I, E, C0, L_min):
    n, T = P.shape
    IE = np.array(I, dtype=float) - np.array(E, dtype=float)
    pagos_sem = P.sum(axis=0)
    C = C0 + np.cumsum(IE) - np.cumsum(pagos_sem)
    U = IE - pagos_sem

    filas = []
    for t in range(T):
        fila = {"Semana": t + 1}
        for i, concepto in enumerate(conceptos):
            fila[f"Pagar {concepto} ($)"] = round(float(P[i, t]), 2)
        fila["Utilidad/Retenida ($)"] = round(float(U[t]), 2)
        fila["Caja Proyectada ($)"] = round(float(C[t]), 2)
        fila["Ratio D/C"] = round(float(ratios[t]), 4)
        if C[t] <= L_min:
            estado = "Sin margen"
        elif C[t] <= L_min * 1.5:
            estado = "Ajustado"
        else:
            estado = "Operable"
        fila["Estado"] = estado
        filas.append(fila)

    return pd.DataFrame(filas)
```

```python
AGRESIVIDAD = 0.5

P_a, ratios_a, exito_a = modelo5_opcion_a(D0, I, E, C0, L_MIN, HORIZONTE, agresividad=AGRESIVIDAD)
tabla_a = construir_tabla_modelo5(P_a, ratios_a, conceptos, I, E, C0, L_MIN)

print("Modelo 5 - Opción A (heurística mejorada)")
display(tabla_a)
```
