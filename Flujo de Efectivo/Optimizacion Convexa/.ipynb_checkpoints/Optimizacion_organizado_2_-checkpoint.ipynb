# Optimización de Flujo de Caja
---
**Objetivo:** Análisis de ingresos, egresos y deuda semanal con modelo de optimización de pagos.

## 1. Configuración e Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

## 2. Carga de Datos

```python
# Rutas de archivos
PATH_DATE_DEF     = "Tablero de trabajo(Fechas de Semanas).csv"
PATH_CONCEPT_CLASQ = "Tablero de trabajo(Clasificacion de Conceptos).csv"
PATH_CASH_FLOW    = "Tablero de trabajo(Flujos de E).csv"
```

```python
# Fechas de semanas
df_dates = pd.read_csv(PATH_DATE_DEF, encoding='latin-1', sep=';', quotechar='"')
df_dates
```

```python
# Flujo de efectivo (cash flow)
df_tmp   = pd.read_csv(PATH_CASH_FLOW, encoding='latin-1', sep=';', quotechar='"', header=None)
conceptos = df_tmp.iloc[0, 1:]
datos     = df_tmp.iloc[1:, 1:]
semanas   = df_tmp.iloc[1:, 0]

df_cash_flow = pd.DataFrame(datos.values, columns=conceptos.values, index=semanas.values)
df_cash_flow = df_cash_flow.T
df_cash_flow
```

```python
# Clasificación de conceptos
df_concepts = pd.read_csv(PATH_CONCEPT_CLASQ, encoding='latin-1', sep=';', quotechar='"')
df_concepts
```

## 3. Limpieza y Preprocesamiento

```python
# 3.1 Renombrar columnas duplicadas
duplicados = df_cash_flow.columns[df_cash_flow.columns.duplicated()].unique()
if len(duplicados) > 0:
    print("Columnas duplicadas encontradas:", duplicados)
    cols = df_cash_flow.columns.to_series()
    df_cash_flow.columns = cols.where(
        ~cols.duplicated(),
        cols + '_' + cols.groupby(cols).cumcount().add(1).astype(str)
    )
else:
    print("Sin columnas duplicadas.")

# 3.2 Limpiar valores monetarios (quitar $, puntos de miles, etc.)
for col in df_cash_flow.columns:
    df_cash_flow[col] = pd.to_numeric(
        df_cash_flow[col]
        .astype(str)
        .str.replace(r'[$\s]', '', regex=True)
        .str.replace(r'\.(?=\d{3})', '', regex=True)   # punto de miles
        .str.replace(',', '.'),
        errors='coerce'
    )

print("Columnas disponibles:")
for c in df_cash_flow.columns:
    print(" ", repr(c))
```

## 4. Definición de Conceptos

```python
# ── AJUSTÁ ESTAS LISTAS SEGÚN TUS COLUMNAS ──────────────────────────────────
conceptos_ingresos = [
    ' Proberry ',
    ' Reimer ',
    ' GALU ',
    ' Efectivo ',
    ' Saldo LAR ',
    ' Prestamo Embolsadora ',
    ' Pago AVV ',
    ' Mauricio Caro ',
    ' Pago Luinni ',
    ' Crumble otros ',
]

conceptos_egresos = [
    ' AVV ',
    ' Aceros Gutz ',
    ' Arturo Guerrero ',
    ' BATAS, DIESEL, HOMEDEPOT ',
    ' Reimer ',
    ' GALU ',
    ' Efectivo ',
    ' Saldo LAR ',
    ' Prestamo Embolsadora ',
    ' Pago AVV ',
    ' Mauricio Caro ',
    ' Pago Luinni ',
    ' Crumble otros ',
]
# ─────────────────────────────────────────────────────────────────────────────

# Filtrar solo las columnas que existen en el DataFrame
ingresos_cols = [c for c in conceptos_ingresos if c in df_cash_flow.columns]
egresos_cols  = [c for c in conceptos_egresos  if c in df_cash_flow.columns]

print(f"Ingresos encontrados : {len(ingresos_cols)} / {len(conceptos_ingresos)}")
print(f"Egresos encontrados  : {len(egresos_cols)} / {len(conceptos_egresos)}")
```

```python
# Totales semanales
ingreso_semanal = df_cash_flow[ingresos_cols].sum(axis=1, skipna=True)
egreso_semanal  = df_cash_flow[egresos_cols].sum(axis=1,  skipna=True)

df_ingresos = ingreso_semanal.reset_index()
df_ingresos.columns = ['Semana', 'Ingreso']

df_egresos = egreso_semanal.reset_index()
df_egresos.columns = ['Semana', 'Egreso']

# Flujo neto
flujo_neto = ingreso_semanal - egreso_semanal
df_flujo = pd.DataFrame({
    'Semana' : flujo_neto.index,
    'Ingreso': ingreso_semanal.values,
    'Egreso' : egreso_semanal.values,
    'Neto'   : flujo_neto.values,
})

print(f"""{df_ingresos.to_string(index=False, formatters={'Ingreso': '${:,.2f}'.format})}, 
      {df_egresos.to_string(index=False, formatters={'Ingreso': '${:,.2f}'.format})}, 
      {df_flujo.to_string(index=False, formatters={'Ingreso': '${:,.2f}'.format})}""")
```

## 5. Análisis de Deuda

```python
columnas_resp = [c for c in df_cash_flow.columns if c.startswith(' Responsabilidad')]

if not columnas_resp:
    print("No se encontraron columnas de Responsabilidad.")
else:
    df_resp   = df_cash_flow[columnas_resp]
    totales   = df_resp.sum(axis=0, skipna=True)
    total_gral = totales.sum()
    porcentajes = (totales / total_gral) * 100

    resultado = pd.DataFrame({
        'Concepto'   : totales.index,
        'Total'      : totales.values,
        'Porcentaje' : porcentajes.values,
    })
    resultado = resultado[resultado['Total'] > 0].sort_values('Porcentaje', ascending=False)

    print(f"TOTAL GENERAL (Responsabilidad): ${total_gral:,.2f}")
    print(resultado.to_string(index=False, formatters={
        'Total'     : lambda x: f'${x:,.2f}',
        'Porcentaje': lambda x: f'{x:.4f}%',
    }))
```

## 7. Modelo de Optimización — Minimización de Varianza de Caja

### Formulación

$$
\min_{p_{j,t}} \;\alpha \sum_{t=1}^{T}(C_t - \bar{C})^2 + \beta \sum_j D_{j,T}
$$

**Restricciones:**

| # | Restricción | Expresión |
|---|-------------|-----------|
| 1 | Liquidez mínima | $C_t \geq L_{\min}$ |
| 2 | No negatividad de caja | $C_t \geq 0$ |
| 3 | Dinámica de caja | $C_t = C_{t-1} + I_t - E_t - \sum_j p_{j,t}$ |
| 4 | Dinámica de deuda | $D_{j,t} = D_{j,t-1} - p_{j,t}$ |
| 5 | Rango de pagos | $0 \leq p_{j,t} \leq D_{j,t-1}$ |
| 6 | No negatividad de deuda | $D_{j,t} \geq 0$ |
| 7 | Pagos ≤ efectivo disponible | $\sum_j p_{j,t} \leq C_{t-1} + I_t - E_t - L_{\min}$ |

**Parámetros:**
- $\bar{C}$: meta de caja (setpoint) — **definido por el usuario**
- $L_{\min}$: liquidez mínima obligatoria — **definido por el usuario**
- $\alpha$: peso de la varianza de caja — **calibrado automáticamente** como $1/\text{Var}(\text{flujo histórico})$
- $\beta$: peso de la deuda residual — **calibrado automáticamente** como $1/D_{\text{total}}$

### 7.1 Parámetros del Modelo

```python
# ┌─────────────────────────────────────────────────────────┐
# │   ÚNICOS VALORES QUE EL USUARIO NECESITA CAMBIAR        │
# └─────────────────────────────────────────────────────────┘

C_BAR  = 400_000   # Meta de caja (setpoint): ¿cuánto quieres tener en caja?
L_MIN  = 200_000   # Liquidez mínima obligatoria (nunca bajar de este nivel)

# ── Horizonte de proyección ───────────────────────────────────────────────────
T_HORIZON = 8      # semanas hacia adelante a optimizar

# ── Caja inicial (dato real, no modificar) ───────────────────────────────────
# Se toma como la posición acumulada de flujo neto histórico.
# Si tienes el saldo real de caja, reemplaza con: C_INIT = <tu_valor>
C_INIT = float(df_flujo['Neto'].cumsum().iloc[-1])

# ── Acreedores y saldos de deuda ─────────────────────────────────────────────
# Se toman automáticamente de las columnas Responsabilidad* (Sección 5).
# Si no existen, se puede definir manualmente abajo.
if 'df_resp' in dir() and df_resp is not None and len(df_resp.columns) > 0:
    nombres_deuda = list(df_resp.columns)
    D0 = df_resp.mean(axis=0).values.astype(float)
else:
    # Fallback manual
    nombres_deuda = ['Acreedor_1', 'Acreedor_2', 'Acreedor_3']
    D0 = np.array([500_000, 300_000, 200_000], dtype=float)

J = len(D0)

print("── Configuración del modelo ──────────────────────────────────────")
print(f"  Meta de caja  (C̄)      : ${C_BAR:>15,.0f}  ← definido por usuario")
print(f"  Liquidez mínima (L_min) : ${L_MIN:>15,.0f}  ← definido por usuario")
print(f"  Caja inicial  (C₀)      : ${C_INIT:>15,.0f}  ← dato real")
print(f"  Horizonte               : {T_HORIZON} semanas")
print(f"  Nro. de acreedores      : {J}")
print()
for j, (nom, saldo) in enumerate(zip(nombres_deuda, D0)):
    print(f"    [{j+1}] {nom:<35}: ${saldo:>12,.0f}")
```

### 7.2 Calibración Automática de Parámetros α y β

```python
import cvxpy as cp
import openpyxl
from openpyxl.styles import PatternFill, Font, Alignment, Border, Side
from openpyxl.utils import get_column_letter

# Alpha: normalizado por varianza histórica del flujo neto
# → controla cuánto penaliza la desviación de caja respecto a C̄
var_flujo_hist = np.var(df_flujo['Neto'].values)
ALPHA = 1.0 / var_flujo_hist if var_flujo_hist > 0 else 1.0

# Beta: normalizado por deuda total
# → controla cuánto penaliza la deuda residual al final del horizonte
deuda_total = D0.sum()
BETA = 1.0 / deuda_total if deuda_total > 0 else 1.0

print("── Parámetros calibrados automáticamente ─────────────────────────")
print(f"  Varianza histórica flujo : ${var_flujo_hist:>15,.0f}")
print(f"  ALPHA  (peso varianza)   :  {ALPHA:.4e}  ← 1 / Var(flujo histórico)")
print(f"  BETA   (peso deuda)      :  {BETA:.4e}  ← 1 / Deuda total")
print()
print("  Tip: sube ALPHA para priorizar estabilidad de caja.")
print("  Tip: sube BETA  para liquidar deuda más agresivamente.")
```

### 7.3 Proyección de Flujos

```python
# Proyección por promedio móvil de las últimas 4 semanas.
# Para usar proyecciones propias, reemplaza I_proj y E_proj
# con arrays numpy de longitud T_HORIZON.
ventana = 4

I_proj = np.full(T_HORIZON, df_flujo['Ingreso'].iloc[-ventana:].mean())
E_proj = np.full(T_HORIZON, df_flujo['Egreso'].iloc[-ventana:].mean())

print("── Proyección de flujos ──────────────────────────────────────────")
print(f"  Base: promedio de las últimas {ventana} semanas")
print(f"  Ingreso semanal proyectado : ${I_proj[0]:>12,.0f}")
print(f"  Egreso  semanal proyectado : ${E_proj[0]:>12,.0f}")
print(f"  Flujo neto proyectado      : ${I_proj[0]-E_proj[0]:>12,.0f}")
```

### 7.4 Resolución del Modelo (Solver)

```python
# Variables de decisión: p[j, t] = pago al acreedor j en semana t
p = cp.Variable((J, T_HORIZON), nonneg=True)

C_list   = []
D_prev   = D0.copy()
C_prev   = C_INIT
constraints = []

for t in range(T_HORIZON):
    pago_total_t = cp.sum(p[:, t])

    # Dinámica de caja
    C_t = C_prev + I_proj[t] - E_proj[t] - pago_total_t
    C_list.append(C_t)

    # Restricciones de caja
    constraints += [C_t >= L_MIN, C_t >= 0]

    # Pagos no superan efectivo disponible tras gastos y L_min
    efectivo_disp = C_prev + I_proj[t] - E_proj[t] - L_MIN
    constraints  += [pago_total_t <= efectivo_disp]

    # Restricciones por acreedor
    for j in range(J):
        constraints += [p[j, t] <= D_prev[j]]

    # Avanzar estado
    C_prev = C_t
    D_prev = [D_prev[j] - p[j, t] for j in range(J)]

# Deuda residual al final del horizonte
D_final_expr = cp.sum([D_prev[j] for j in range(J)])

# Función objetivo: min α·Σ(Cₜ-C̄)² + β·Σ Dⱼ,T
varianza_caja = cp.sum([(C_list[t] - C_BAR)**2 for t in range(T_HORIZON)])
objetivo      = ALPHA * varianza_caja + BETA * D_final_expr

prob = cp.Problem(cp.Minimize(objetivo), constraints)
prob.solve(solver=cp.CLARABEL, verbose=False)

print(f"── Solver ────────────────────────────────────────────────────────")
print(f"  Estado    : {prob.status}")
print(f"  Valor obj.: {prob.value:.6f}" if prob.value is not None else "  Valor obj.: N/A")

if prob.status not in ["optimal", "optimal_inaccurate"]:
    print()
    print("⚠️  Sin solución óptima. Verifica que:")
    print("   · C_INIT y flujos proyectados cubran L_MIN")
    print("   · L_MIN no sea mayor que el flujo neto esperado")
    print("   · Considera reducir C_BAR o L_MIN")
```

### 7.5 Tabla de Recomendaciones

```python
if prob.status in ["optimal", "optimal_inaccurate"]:
    p_val = p.value
    C_val = np.array([float(c.value) for c in C_list])

    # Trayectoria de deuda
    D_tray = [D0.copy()]
    for t in range(T_HORIZON):
        D_tray.append(D_tray[-1] - p_val[:, t])

    # Construir DataFrame de recomendaciones
    filas = []
    for t in range(T_HORIZON):
        caja    = C_val[t]
        ratio   = caja / L_MIN
        semaforo = "🟢 OK"      if ratio > 1.5 else \
                   "🟡 Cuidado" if ratio > 1.0 else \
                   "🔴 Riesgo"
        alerta   = ""                       if ratio > 1.5 else \
                   "Caja cerca del mínimo"  if ratio > 1.0 else \
                   "¡Riesgo de liquidez!"

        fila = {
            'Semana'         : t + 1,
            'Caja Proyectada': caja,
            'vs Meta (C̄)'   : caja - C_BAR,
            'Semáforo'       : semaforo,
            'Alerta'         : alerta,
        }
        for j, nom in enumerate(nombres_deuda):
            fila[f'Pagar {nom}'] = p_val[j, t]
        filas.append(fila)

    df_recomendaciones = pd.DataFrame(filas)

    # Mostrar tabla
    cols_fmt = {
        'Caja Proyectada': '${:>13,.0f}'.format,
        'vs Meta (C̄)'   : lambda x: f'${x:>+13,.0f}',
    }
    for nom in nombres_deuda:
        cols_fmt[f'Pagar {nom}'] = '${:>13,.0f}'.format

    print("── Recomendaciones por Semana ─────────────────────────────────\n")
    print(df_recomendaciones.to_string(index=False, formatters=cols_fmt))

    # Resumen de deuda
    print("\n── Deuda al final del horizonte ───────────────────────────────")
    for j, nom in enumerate(nombres_deuda):
        pagado   = p_val[j, :].sum()
        restante = D_tray[-1][j]
        print(f"  {nom:<35}: pagado ${pagado:>12,.0f}  |  restante ${restante:>12,.0f}")
    print(f"  {'TOTAL':<35}: pagado ${p_val.sum():>12,.0f}  |  restante ${D_tray[-1].sum():>12,.0f}")
else:
    print("⚠️  No hay recomendaciones disponibles: el solver no encontró solución óptima.")
```

### 7.6 Exportar a Excel para Administración

```python
if prob.status in ["optimal", "optimal_inaccurate"]:
    PATH_OUTPUT = "Recomendaciones_Caja.xlsx"

    wb = openpyxl.Workbook()
    ws = wb.active
    ws.title = "Recomendaciones"

    # Estilos
    VERDE    = PatternFill("solid", fgColor="C6EFCE")
    AMARILLO = PatternFill("solid", fgColor="FFEB9C")
    ROJO     = PatternFill("solid", fgColor="FFC7CE")
    GRIS     = PatternFill("solid", fgColor="D9D9D9")
    AZUL     = PatternFill("solid", fgColor="BDD7EE")

    borde_lado = Side(style="thin", color="AAAAAA")
    borde = Border(left=borde_lado, right=borde_lado, top=borde_lado, bottom=borde_lado)

    def celda(ws, fila, col, valor, fill=None, font=None, alin=None, fmt=None):
        c = ws.cell(row=fila, column=col, value=valor)
        if fill : c.fill           = fill
        if font : c.font           = font
        if alin : c.alignment      = alin
        if fmt  : c.number_format  = fmt
        c.border = borde
        return c

    centro  = Alignment(horizontal="center", vertical="center")
    derecha = Alignment(horizontal="right")
    FMT_M   = '$#,##0'

    # Título
    ws.merge_cells(f"A1:{get_column_letter(5 + J)}1")
    c = ws.cell(row=1, column=1, value="Recomendaciones de Pago y Proyección de Caja")
    c.font = Font(bold=True, size=13); c.alignment = centro; c.fill = AZUL

    ws.merge_cells(f"A2:{get_column_letter(5 + J)}2")
    c = ws.cell(row=2, column=1,
                value=f"Horizonte: {T_HORIZON} sem  |  Meta caja: ${C_BAR:,.0f}  |  Liquidez mín: ${L_MIN:,.0f}")
    c.alignment = centro; c.font = Font(italic=True, size=9)

    # Encabezados
    encabezados = ['Semana', 'Caja Proyectada ($)', 'vs Meta ($)', 'Semáforo', 'Alerta']
    for nom in nombres_deuda:
        encabezados.append(f'Pagar {nom} ($)')

    for col, enc in enumerate(encabezados, 1):
        celda(ws, 4, col, enc, fill=GRIS, font=Font(bold=True, size=10), alin=centro)

    # Datos
    for i, row in df_recomendaciones.iterrows():
        fila  = i + 5
        ratio = row['Caja Proyectada'] / L_MIN
        color = VERDE if ratio > 1.5 else (AMARILLO if ratio > 1.0 else ROJO)

        celda(ws, fila, 1, int(row['Semana']),       alin=centro)
        celda(ws, fila, 2, row['Caja Proyectada'],   fill=color,  fmt=FMT_M,            alin=derecha)
        celda(ws, fila, 3, row["vs Meta (C̄)"],      fmt='$#,##0;($#,##0)',              alin=derecha)
        celda(ws, fila, 4, row['Semáforo'],          alin=centro)
        celda(ws, fila, 5, row['Alerta'],            alin=centro)

        for j, nom in enumerate(nombres_deuda):
            val = row[f'Pagar {nom}']
            celda(ws, fila, 6 + j, val,
                  fill=VERDE if val > 0 else None, fmt=FMT_M, alin=derecha)

    # Fila totales
    fila_tot = T_HORIZON + 5
    for col in range(1, 6):
        celda(ws, fila_tot, col, "TOTAL PAGOS" if col == 1 else "", fill=GRIS, font=Font(bold=True), alin=centro)
    for j in range(J):
        celda(ws, fila_tot, 6 + j, p_val[j].sum(),
              fill=GRIS, font=Font(bold=True), fmt=FMT_M, alin=derecha)

    # Leyenda
    fila_ley = fila_tot + 2
    ws.cell(row=fila_ley, column=1, value="Leyenda:").font = Font(bold=True)
    for col, (txt, fill) in enumerate([
        ("🟢  Caja > 1.5× mínimo",           VERDE),
        ("🟡  Caja entre 1× y 1.5× mínimo",  AMARILLO),
        ("🔴  Caja en riesgo",                ROJO),
    ], 2):
        c = ws.cell(row=fila_ley, column=col, value=txt)
        c.fill = fill; c.alignment = centro; c.border = borde

    # Ancho columnas
    for i, ancho in enumerate([10, 22, 18, 14, 30] + [22]*J, 1):
        ws.column_dimensions[get_column_letter(i)].width = ancho

    ws.row_dimensions[1].height = 22
    ws.row_dimensions[4].height = 18
    ws.freeze_panes = "A5"

    wb.save(PATH_OUTPUT)
    print(f"✅ Excel exportado: {PATH_OUTPUT}")
    print("   Compártelo con el área de administración.")
else:
    print("⚠️  No se generó Excel: el solver no encontró solución óptima.")
```
