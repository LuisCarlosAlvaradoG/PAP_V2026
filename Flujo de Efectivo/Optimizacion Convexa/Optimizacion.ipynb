#### RevisiÃ³n de la Deuda

```python
def convertir_moneda(valor):
    if pd.isna(valor):
        return np.nan
    if isinstance(valor, (int, float)):
        return float(valor)
    try:
        limpio = str(valor).replace('$', '').replace('.', '').replace(',', '.')
        return float(limpio)
    except:
        return np.nan

columna_conceptos = df_cash_flow.columns[0]
semanas = df_cash_flow.columns[1:].tolist()

mask = df_cash_flow[columna_conceptos].str.strip().str.startswith('Responsabilidad', na=False)
df_resp = df_cash_flow.loc[mask].copy()

if df_resp.empty:
    print("â ï¸ No se encontraron filas que empiecen con 'Responsabilidad'.")
    print("ð¡ Revisa los primeros valores de la columna:")
    print(df_cash_flow[columna_conceptos].head(10).tolist())
else:
    for col in semanas:
        if col in df_resp.columns:
            df_resp[col] = df_resp[col].apply(convertir_moneda)

    df_resp['Total_Concepto'] = df_resp[semanas].sum(axis=1, skipna=True)
    total_general = df_resp['Total_Concepto'].sum()
    df_resp['Porcentaje'] = (df_resp['Total_Concepto'] / total_general) * 100

    print("ð° TOTAL GENERAL DE DEUDA (solo Responsabilidad)")
    print("=" * 60)
    print(f"$ {total_general:,.2f}".replace(',', '.').replace('.', ',', 1))

    print("\n" + "=" * 60)
    print("ð DETALLE POR CONCEPTO (solo con deuda > 0, ordenado por % DESC)")
    print("=" * 60)

    resultado = df_resp[[columna_conceptos, 'Total_Concepto', 'Porcentaje']]

    resultado = resultado[resultado['Total_Concepto'].notna() & (resultado['Total_Concepto'] > 0)]

    resultado = resultado.sort_values('Porcentaje', ascending=False)

    if resultado.empty:
        print("â ï¸ No hay conceptos con deuda > 0.")
    else:
        print(resultado.to_string(index=False, formatters={
            'Total_Concepto': lambda x: f"$ {x:,.2f}".replace(',', '.').replace('.', ',', 1),
            'Porcentaje': lambda x: f"{x:.4f} %"
        }))
```

#### Modelo Propuesto

$
\min_{t} \sum \left[
\sum_{j} x_{j} D_{j,t}
- \alpha U_{t}
+ \beta \max(0, L_{min} - C_{t})
\right]
$

##### Restricciones/Sujeto a:
* _Liquidez_: $C_t â¥ L_{min}$
* _No negatividad de caja_: $C_t \geq 0$
* _DinÃ¡mica de caja_:$C_t = C_{t-1} + U_t - \sum_j p_{j,t}$
* _DinÃ¡mica de deuda_:$D_{j,t} = D_{j,t-1} - p_{j,t}$
* _RestricciÃ³n de pagos por deuda_:$0 \leq p_{j,t} \leq D_{j,t}$
* No negatividad de deuda_:$D_{j,t} \geq 0$
* RestricciÃ³n de pagos:$\sum_j p_{j,t} \leq C_t$
